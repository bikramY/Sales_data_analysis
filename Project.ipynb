```python
#checking order placement data for potential patters or trends in  purchasing 
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#females are buying more than males
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum()
sales_age
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)#sorting the values
sales_age
```

```python
#age group 26-35 is the most potential customers
```

```python
ax=sns.barplot(x="Age Group",y="Amount",data=sales_age)
```

```python
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
ax=sns.barplot(x="State",y="Orders",data=sales_state)
```

```python
#utter pradesh maharstra are potential customers
```

```python
#checking for maritial status
ax=sns.countplot(x="Marital_Status",data=df)
sns.set(rc={"figure.figsize":(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#checking on occupation
ax=sns.countplot(x="Occupation",data=df)
sns.set(rc={"figure.figsize":(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# IT,Healthcare,Aviation
```

```python
#checking on occupation

sales_state = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by ="Amount", ascending = False).head(8)

sns.set(rc={"figure.figsize":(15,3)})
ax = sns.barplot(x= "Product_Category", y = "Amount", data= sales_state)
```

```python
#top potential customers Occupation-wise -----IT, Healthcare, Aviation
#top potential customers gender-wise----------females are likely to buy almost twice more than that of males
#top potential customers age-wise-------------age group 26-35 is the best targer group 
#top potential customers state-wise---------- uttar pradesh and maharasta
#top potential customers product_category-wise--------food 
```
